### Cleaning 'Post_Time' Column / Creating New Column ('Days_Since_Post') 

```python
df_final['Post_Time'].unique()
```

```python
def post_time_num(p):
    if p['Post_Time'] == 'Hiring ongoing':
        return 'Ongoing'
    elif p['Post_Time'] == 'Posted\nToday' or p['Post_Time'] == 'Posted\nJust posted': 
        return 0
    else: 
        return p['Post_Time']
        
df_final['Post_Time'] = df_final.apply(post_time_num, axis = 1)
```

```python
df_final['Post_Time'] = df_final['Post_Time'].astype(str)
```

```python
df_final['Days_Since_Post'] = df_final['Post_Time'].str.extract('(\d+)')
```

```python
df_final['Days_Since_Post'] = df_final['Days_Since_Post'].fillna('Ongoing')
```

```python
df_final.loc[df_final['Post_Time'].str.contains("30+", na=df_final['Days_Since_Post']), "Days_Since_Post"] = "30+"
```

```python
df_final.sample(15)
```
